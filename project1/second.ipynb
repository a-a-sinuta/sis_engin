```python
import pandas as pd

# Создание объекта DataFrame
df = pd.DataFrame({
    'Товар': ['Молоко', 'Хлеб', 'Сыр', 'Мясо', 'Овощи'],
    'Количество': [10, 5, 15, 20, 12],
    'Цена': [50, 30, 70, 90, 60]
})

df
```

```python
# Создание сводной таблицы
pivoted = df.pivot_table(index='Товар', columns='Количество', values='Цена', aggfunc='sum')

# Вывод сводной таблицы
print(pivoted)
```
